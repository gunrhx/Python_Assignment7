```python
#importing packages
import pandas as pd
import numpy as np

rng = np.random.default_rng()

#### CREATE DATAFRAME ####
# 1. Assign random variables
df = pd.DataFrame({
    "subject":  range(1, 7),
    "gender": rng.choice(["Male", "Female"], size = 6),
    "age": rng.integers(15, 41, 6),
    "depression": rng.integers(0, 2, 6)
})

# 2. Save dataframe
df.to_csv("random_data.csv", index = False)
```

```python
#### EDITING FOR ASSIGNMENT 7 ####

import matplotlib.pyplot as plt

# 1. Checking for na in df
df.isna().sum()

# 2. Descriptive data
df.describe()

# 3. Plotting depression
plt.scatter(df['age'], df['depression'])
plt.xlabel('Age')
plt.ylabel('Depression')
plt.title('Depression by Age')
```
